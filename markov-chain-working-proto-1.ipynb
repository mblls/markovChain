after loading in our netflix data, let's preprocess it with some pandas

```python
df = pd.DataFrame(d)
```

```python
train_data, train_label = df['title'].iloc[0:4500], df['description'].iloc[0:4500]
test_data, test_label =  df['title'].iloc[4501:6233], df['description'].iloc[4501:6233]

train_data, train_label = np.array(train_data), np.array(train_label)
test_data, test_label =  np.array(test_data), np.array(test_label)
```

Now we can create out markov chain.

```python
import random
index = 1
chain = {}
count = 10

title_array = []
for sentence in train_label:
    title_words = sentence.split(' ')
    title_array.extend(title_words)


for word in title_array[index:]:
    key = title_array[index-1]
    if key in chain:
        chain[str(key)].append(str(word))
    else:
        chain[str(key)] = [str(word)]
    index += 1

#print(chain)

word1 = random.choice(list(chain.keys()))
message = word1.capitalize()

while len(message.split(' ')) < count:
    #print(chain[word1])
    word2 = random.choice(chain[word1])

    word1 = word2
    message += ' ' + str(word2)


print(message)
```
